```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style='darkgrid', palette='colorblind', \
        color_codes=True)

cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target
X_train = X[:450]
X_test = X[450:]
y_train = Y[:450] #target
y_test= Y[450:] #target 
```

```python
class Perceptron:

    def __init__(self, learning_rate, n_iters):
        self.learning_rate = learning_rate
        self.n_iters = n_iters
        self.activation_func = self._unit_step_func
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_rows, n_columns = X.shape

        # init parameters
        self.weights = np.zeros(n_columns)
        self.bias = 0

        #making sure that the values passed for y are only 1s or 0s 
        #y_ = np.array([1 if i > 0 else 0 for i in y]) 

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)
                
                # Perceptron update rule
                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = self.activation_func(linear_output)
        return y_predicted

    def _unit_step_func(self, x):
        return np.where(x>=0, 1, 0)
```

```python
def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy
```

```python
p = Perceptron(learning_rate=0.0001, n_iters=1000)
p.fit(X_train, y_train)
predictions = p.predict(X_test)

print("Perceptron classification accuracy", \
      accuracy(y_test, predictions))

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.scatter(X_train[:,0], X_train[:,1],marker='x',c=y_train)

x0_1 = np.amin(X_train[:,0])
x0_2 = np.amax(X_train[:,0])

x1_1 = (-p.weights[0] * x0_1 - p.bias) / p.weights[1]
x1_2 = (-p.weights[0] * x0_2 - p.bias) / p.weights[1]

ax.plot([x0_1, x0_2],[x1_1, x1_2], 'k')

ymin = np.amin(X_train[:,1])
ymax = np.amax(X_train[:,1])
ax.set_ylim([ymin-3,ymax+3])

plt.show()
```

**Q1. Let us modify the update rule to add a learning rate α which multiplies the update
applied to weights. The update rule becomes:**

$$
w' \leftarrow w + \alpha t^{(i)}x^{(i)}
$$
**Show that the value of the
learning rate does not change the perceptron’s prediction after an update.**

```python
learning_rates = [1,0.1, 0.001, 0.0001, 0.00001]
for i in range(len(learning_rates)-1):
  print(learning_rates[i]) 
  p = Perceptron(learning_rate=learning_rates[i], n_iters=1000)
  p.fit(X_train, y_train)
  predictions = p.predict(X_test)
  print("Learning rate:{}, Accuracy:{}".\
        format(learning_rates[i],accuracy(y_test, predictions)))
```
